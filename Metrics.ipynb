```python
metrics = defaultdict()
metrics['train accuracy'] = dict()
metrics['train total loss'] = dict()
metrics['test accuracy'] = dict()
metrics['test fscore'] = dict()
metrics['test precision'] = dict()
metrics['test recall'] = dict()

i = 0
for pair in hyperparameters:
  batch,epoch,lr = pair
  metrics['train accuracy'][tuple(pair)] = result[i][0]
  metrics['train total loss'][tuple(pair)] = result[i][1]
  metrics['test accuracy'][tuple(pair)] =result[i][2]
  metrics['test fscore'][tuple(pair)] = result[i][3]
  metrics['test precision'][tuple(pair)] = result[i][4]
  metrics['test recall'][tuple(pair)] = result[i][5]
  i+=1
```

```python
def plot_entrie(metric):
  y = [value for value in metrics[metric].values()]
  y = np.array(y)
  x = [str(key) for key in metrics[metric].keys()]
  plt.bar(x,y )
  plt.xticks(x,rotation='vertical')
  plt.ylim(0.95,1)
  plt.xlabel("Parameter set")
  plt.ylabel(metric)
  plt.title("Overview on %s"%(metric))
  plt.show()

plot_entrie('train accuracy')
plot_entrie('test accuracy')
plot_entrie('test fscore')
```
